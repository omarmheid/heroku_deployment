```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from functools import reduce
from sqlalchemy import create_engine
```

```python
#Files Used
#01 countries of the world.csv
#02 Travel and Tourism direct contribution to GDP.xlsx
#03 self_employed_percent_of_employment.xlsx
#04 suicide-rates-by-country.xlsx
#05 List_of_countries_by_antidepressant_consumption.xlsx
#06 share-population-female.csv
#07 Household Size and Composition.csv
```

```python
#File01=Countries of the World
f1_path= "countries of the world.csv"
t1_read = pd.read_csv(f1_path, delimiter=',')
#Create dataframe and format
country_stats = pd.DataFrame({
                            'Country':t1_read['Country'],
                            #'Country_Code':"",
                            #'Year':0,
                            'PopDenisty_per_mile':t1_read['Pop. Density (per sq. mi.)'],
                            'Coast_Line_Ratio':t1_read['Coastline (coast/area ratio)'],
                            'PhoneUse_per_1k':t1_read['Phones (per 1000)'],
                            'Birthrate':t1_read['Birthrate']
})

country_stats['PopDenisty_per_mile_Ranked'] = country_stats['PopDenisty_per_mile'].rank(ascending=1)
country_stats['Coast_Line_Ratio_Ranked'] = country_stats['Coast_Line_Ratio'].rank(ascending=0)
country_stats['PhoneUse_per_1k_Ranked'] = country_stats['PhoneUse_per_1k'].rank(ascending=1)
country_stats['Birthrate_Ranked'] = country_stats['Birthrate'].rank(ascending=0)

country_stats
```

```python
#File02=Travel and Tourism direct contribution to GDP focusing on Percent growth of in Country Travel for 2019
f2_path= "Travel and Tourism direct contribution to GDP.xlsx"
intourism = pd.DataFrame(pd.read_excel(f2_path, sheet_name='Travel and Tourism direct contr'))

intourism_cln1=intourism['Indicator']=='Internal Travel and Tourism consumption'
intourism_cln2=intourism[intourism_cln1]
intourism_cln3=intourism_cln2['Subindicator Type']=="% growth"
intourism_cln4=intourism_cln2[intourism_cln3]

#Create dataframe and format
incountry_tourism = pd.DataFrame({
                            'Country':intourism_cln4['Country Name'],
                            #'Country_Code':intourism_cln4['Country ISO3'],
                            #'Year':2019,
                            'Percent_incountry_Travel_Growth':intourism_cln4[2019]})
                            
incountry_tourism['Percent_incountry_Travel_Growth_Ranked'] = incountry_tourism['Percent_incountry_Travel_Growth'].rank(ascending=0)
incountry_tourism = incountry_tourism.reset_index(drop=True)
pd.set_option('display.max_rows', 200)
incountry_tourism
```

```python
#File03=Percent self employeed by country
f3_path= "self_employed_percent_of_employment.xlsx"
self_emplyed = pd.DataFrame(pd.read_excel(f3_path, sheet_name='self_employed_percent_of_employ'))

#Create dataframe and format
self_emplyed = pd.DataFrame({
                            'Country':self_emplyed['country'],
                            #'Country_Code':"",
                            #'Year':2019,
                            'Percent_self_emplyed':self_emplyed[2019]})
                            
self_emplyed['Percent_self_emplyed_Ranked'] = self_emplyed['Percent_self_emplyed'].rank(ascending=0)
self_emplyed = self_emplyed.reset_index(drop=True)
pd.set_option('display.max_rows', 200)
self_emplyed
```

```python
#File04=suicide rate (age-adjusted suicides per 100,000 people)
f4_path= "suicide-rates-by-country.xlsx"
suicide_by_country_df = pd.DataFrame(pd.read_excel(f4_path, sheet_name='suicide-rates-by-country'))

suicide_by_country_all=pd.DataFrame({
                            'Country':suicide_by_country_df['Entity'],
                            #'Country_Code':suicide_by_country_df['Code'],
                            #'Year':0,
                            'suicide_per_100k':suicide_by_country_df['suicide rate (age-adjusted suicides per 100,000 people)']})

suicide_by_country = suicide_by_country_all.groupby(['Country']).agg({'suicide_per_100k': 'mean'}).reset_index()

suicide_by_country['suicide_per_100k_Ranked'] = suicide_by_country['suicide_per_100k'].rank(ascending=1)

pd.set_option('display.max_rows', 200)

suicide_by_country
```

```python
#File05=antidepressant consumption-Defined daily dosage per 1 000 inhabitants per day
f5_path= "List_of_countries_by_antidepressant_consumption.xlsx"
antidepressant_df = pd.DataFrame(pd.read_excel(f5_path, sheet_name='antidepressant consumption'))

#Create dataframe and format
antidepressant_use = pd.DataFrame({
                            'Country':antidepressant_df['Country'],
                            #'Country_Code':"",
                            #'Year':0,
                            'antidepressant_per_1k':antidepressant_df['antidepressant consumption-Defined daily dosage per 1 000 inhabitants per day']})
                            
antidepressant_use['antidepressant_per_1k_Ranked'] = antidepressant_use['antidepressant_per_1k'].rank(ascending=1)

pd.set_option('display.max_rows', 200)
antidepressant_use
```

```python
#File06=Population, female (% of total) (% of total)
f6_path= "share-population-female.csv"
t6_read = pd.read_csv(f6_path, delimiter=',')

Per_of_FemalePop1=t6_read['Year']==2017
Per_of_FemalePop2=t6_read[Per_of_FemalePop1]

Per_of_FemalePop = pd.DataFrame({
                            'Country':Per_of_FemalePop2['Entity'],
                            #'Country_Code':Per_of_FemalePop2['Entity'],
                            #'Year':Per_of_FemalePop2['Year'],
                            'Per_of_FemalePop':Per_of_FemalePop2['Population, female (% of total) (% of total)']
                                })
Per_of_FemalePop['Per_of_FemalePop_Ranked'] = Per_of_FemalePop['Per_of_FemalePop'].rank(ascending=0)
pd.set_option('display.max_rows', 200)
Per_of_FemalePop
```

```python
#File07 Household Size and Composition
f7_path= "Household Size and Composition.csv"
t7_read = pd.read_csv(f7_path, delimiter=',',encoding ='latin1',skiprows=1)

Household_all=pd.DataFrame({
                            'Country':t7_read['Country or area'],
                            #'Country_Code':"",
                            #'Year':0,
                            'Household_size':t7_read['Average household size (number of members)'],
                            'Household_Female':t7_read['Female head of household (percentage of households)'],
                            'Household_One':t7_read['One-person'],
                            'Household_Couples_NChildren':t7_read['Couple only'],
                            'Household_Couples_YChildren':t7_read['Couple with children'],
                            'Household_SinglePa_YChildren':t7_read['Single parent with children'],
                            'Household_SingleFe_YChildren':t7_read['Single mother with children'],
                            'Household_SingleMa_YChildren':t7_read['Single father with children'],
                            'Household_ExtFamily':t7_read['Extended family'],
                            'Household_NoRelFamily':t7_read['Non-relatives'],
                            'Household_Unknown':t7_read['Unknown'],
                            'Household_Nuclear':t7_read['Nuclear'],
                            'Household_Multi_generation':t7_read['Multi-generation'],
                            'Household_Three_generation':t7_read['Three generation'],
                            'Household_Skip_generation':t7_read['Skip generation']
                            })

cols = ['Household_size', 'Household_Female', 'Household_One','Household_Couples_NChildren','Household_Couples_YChildren',\
       'Household_SinglePa_YChildren','Household_SingleFe_YChildren','Household_SingleMa_YChildren','Household_ExtFamily',\
       'Household_NoRelFamily','Household_Unknown','Household_Nuclear','Household_Multi_generation',\
        'Household_Three_generation','Household_Skip_generation']
Household_all[cols] = Household_all[cols].apply(pd.to_numeric, errors='coerce', axis=1)

Household=Household_all.groupby(['Country'])[cols].mean().reset_index()

#Rank
Household['Household_size_Ranked'] = Household['Household_size'].rank(ascending=0)
Household['Household_Female_Ranked'] = Household['Household_Female'].rank(ascending=1)
Household['Household_One_Ranked'] = Household['Household_One'].rank(ascending=1)
Household['Household_Couples_NChildren_Ranked'] = Household['Household_Couples_NChildren'].rank(ascending=1)
Household['Household_Couples_YChildren_Ranked'] = Household['Household_Couples_YChildren'].rank(ascending=0)
Household['Household_SinglePa_YChildren_Ranked'] = Household['Household_SinglePa_YChildren'].rank(ascending=1)
Household['Household_SingleFe_YChildren_Ranked'] = Household['Household_SingleFe_YChildren'].rank(ascending=1)
Household['Household_SingleMa_YChildren_Ranked'] = Household['Household_SingleMa_YChildren'].rank(ascending=1)
Household['Household_ExtFamily_Ranked'] = Household['Household_ExtFamily'].rank(ascending=0)
Household['Household_NoRelFamily_Ranked'] = Household['Household_NoRelFamily'].rank(ascending=1)
Household['Household_Unknown_Ranked'] = Household['Household_Unknown'].rank(ascending=1)
Household['Household_Nuclear_Ranked'] = Household['Household_Nuclear'].rank(ascending=0)
Household['Household_Multi_generation_Ranked'] = Household['Household_Multi_generation'].rank(ascending=0)
Household['Household_Three_generation_Ranked'] = Household['Household_Three_generation'].rank(ascending=0)
Household['Household_Skip_generation_Ranked'] = Household['Household_Skip_generation'].rank(ascending=1)

pd.set_option('display.max_rows', 200)

Household
```

```python
data_frames = [country_stats, incountry_tourism, self_emplyed,suicide_by_country,antidepressant_use,Per_of_FemalePop,Household]
#df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Country'],
#                                            how='outer'), data_frames)

# if you want to fill the values that don't exist in the lines of merged dataframe simply fill with required strings as

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Country'],how='outer'), data_frames).fillna('NoData')
pd.set_option('display.max_rows', 200)
df_merged
```

```python
#DataFrames
#1-country_stats
#2-incountry_tourism
#3-self_emplyed
#4-suicide_by_country
#5-antidepressant_use
#6-Per_of_FemalePop
#7-Household

index_a=pd.merge(country_stats, incountry_tourism,  on='Country', how='outer')
index_b=pd.merge(self_emplyed, incountry_tourism,  on='Country', how='outer')
index_c=pd.merge(suicide_by_country, antidepressant_use,  on='Country', how='outer')
index_d=pd.merge(Per_of_FemalePop, Household,  on='Country', how='outer')

happy=pd.concat([index_a, index_b,index_c,index_d])
pd.set_option('display.max_rows', 200)
happy
```
